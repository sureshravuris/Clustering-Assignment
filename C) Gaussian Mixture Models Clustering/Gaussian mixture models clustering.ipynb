```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Clustering by Gaussian Mixture Models
For this notebook and clustering assignment, I will be re-using the actors_meta_df.csv used in the K-Means notebook as it is containing emotional speech data having many dimensions that when clustered together may benefit more from the Gaussian Mixture Model for clustering rather than K-Means clustering.

## Data Loading

```python
actors_meta_df = pd.read_csv('./data_samples/actors_meta_df.csv', on_bad_lines='skip')
```

```python
actors_meta_df.columns
```

```python
numerical_df = actors_meta_df.loc[:, ['modality', 'vocal_channel', 'emotion',
       'emotional_intensity', 'statement', 'repetition']]
```

```python
numerical_df.head()
```

## Data Pre-Processing

```python
# applying the standard scaler to the numerical data
scaler = StandardScaler()
scaler.fit(numerical_df)
scaled_df = scaler.transform(numerical_df)
scaled_df = pd.DataFrame(scaled_df, columns=numerical_df.columns)
scaled_df.head()
```

```python
scaled_df.describe()
```

After scaling, it seems that modality and vocal_channel lack sufficient variance to scale properly and therefore the values have all been scaled to 0 across all values. To preserve interpretability we will just drop these columns (scaled data performs better overall) & re-attempt the analysis, adding a new column from the source data.

```python
scaled_df_interpretable = scaled_df.loc[:, ['emotion', 'emotional_intensity', 'statement', 'repetition']]
scaled_df_interpretable.head()
```

```python
actors_meta_df.tempo.describe()
```

```python
# fit the standard scaler
scaler = StandardScaler()
scaled_tempo = actors_meta_df.tempo.values.reshape(-1, 1)
scaler.fit(scaled_tempo)
scaled_df_interpretable['tempo'] = scaler.transform(scaled_tempo)
```

```python
scaled_df_interpretable.tempo.describe()
```

## Exploratory Data Analysis & Visualization

```python
# applying PCA to the scaled data
pca = PCA(n_components=2)
pca.fit(scaled_df_interpretable)
pca_df = pca.transform(scaled_df_interpretable)
pca_df = pd.DataFrame(pca_df, columns=['PC1', 'PC2'])
pca_df.head()
```

```python
pca_df.shape
```

```python
# plotting the pca
plt.scatter(pca_df['PC1'], pca_df['PC2'])
```

```python
# the cluster centers to use for the GMM model (manually entered and correspond to each of the 12 vertical clusters)
init_means = np.array([(-1.7, 0), (-1.4, 0), (-1.1, 0), (-0.8, 0), (-0.3, 0), (0,0), (0.4, 0), (0.7, 0), (1, 0), (1.4, 0), (1.6, 0), (1.9, 0)])
```

# Clustering
Using Sklearn's Gaussian Mix Model on the pca_df features we created using the scaled audio features 'emotion', 'emotional_intensity', 'statement', 'repetition' and 'tempo' (using 12 components for each vertical line, this time without manually initializing the center and using the kmeans algorithm built into GM)

```python
# fitting the gaussian mixture model
gmm = GaussianMixture(n_components=12)
gmm.fit(pca_df)
gmm_labels = gmm.predict(pca_df)
```

```python
# plot the pca_df using the labels from the gmm
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=gmm_labels)
```

# Assess
The clusters created above do not seem to accurately cluster the very overlapping lines in the center. Increasing max_iter of the GaussianMixture model may improve the clustering but will also increase computation complexity so it is suggested to run this notebook in a remote environment with better CPU/GPU. This time using the init_means I manually created to see if it improves the cluster accuracy. Covariance type is set to 'tied' as each cluster is of similar shape.

```python
# fitting the gaussian mixture model
gmm = GaussianMixture(n_components=12, covariance_type='tied', max_iter = 30000)
gmm.means_init = init_means
#using a smaller sample
pca_df_sample=pca_df.sample(1000)

gmm.fit(pca_df)
gmm_sample_labels = gmm.predict(pca_df_sample)
```

```python
# plot the pca_df using the labels from the gmm
plt.scatter(pca_df_sample['PC1'], pca_df_sample['PC2'], c=gmm_sample_labels)
```

  At first glance, the accuracy of GMM at capturing the 12-15 vertical lines pca clusters looks better than when using the default parameters. My intuition tells me that the 'tempo' column feature (a highly variable and continuous variable) introduced a lot of complexity, as the rest of the features were less variable - more like categoricals - and easily seperable into 2 main clusters. However, I believe this exercise was good practice and a learning experience.

  Each cluster seems to capture 2 of each of the vertical groupings above. Better fine-grained cluster mapping could be achieved with better hyper-parameters, but I believe this is a good start!
